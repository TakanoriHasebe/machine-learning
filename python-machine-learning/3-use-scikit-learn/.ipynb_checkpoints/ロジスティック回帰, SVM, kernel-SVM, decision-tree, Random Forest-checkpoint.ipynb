```python
from sklearn import datasets
import numpy as np
from sklearn.cross_validation import train_test_split
from sklearn.preprocessing import StandardScaler
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
from IPython.display import Image
%matplotlib inline

def versiontuple(v):
    return tuple(map(int, (v.split("."))))

def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.02):

    # setup marker generator and color map
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1],
                    alpha=0.8, c=cmap(idx),
                    marker=markers[idx], label=cl)

    # highlight test samples
    if test_idx:
        # plot all samples
        if not versiontuple(np.__version__) >= versiontuple('1.9.0'):
            X_test, y_test = X[list(test_idx), :], y[list(test_idx)]
            warnings.warn('Please update to NumPy 1.9.0 or newer')
        else:
            X_test, y_test = X[test_idx, :], y[test_idx]

        plt.scatter(X_test[:, 0],
                    X_test[:, 1],
                    c='',
                    alpha=1.0,
                    linewidths=1,
                    marker='o',
                    s=55, label='test set')

# irisデータセットをロード
iris = datasets.load_iris()

# 3, 4列目の特徴量を抽出
X = iris.data[:, [2, 3]]

# クラスラベルを取得
y = iris.target

# 学習データとテストデータに分ける
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# 標準化するために必要
sc = StandardScaler()

# トレーニングデータの平均と標準偏差の計算
sc.fit(X_train)

# 平均と標準偏差を用いて標準化
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)

# 描画するために必要
X_combined_std = np.vstack((X_train_std, X_test_std))
y_combined = np.hstack((y_train, y_test))


from sklearn.svm import SVC
```

## <font color='blue'>ロジスティック回帰</font>

### <font color='blue'>ロジスティック回帰を用いたクラスの確率的モデリング</font>

パーセプトロンの学習の最大の問題点は、クラスを完全に線形分離できない場合は収束しないというところである。<br>
時間を有効に用いるために、ロジスティック回帰について見ていく。<br>

### <font color='blue'>ロジスティック回帰の直観的知識と条件付き確率</font>

ロジスティック回帰は分類モデルであり、非常に実装しやすいが、高い性能が発揮されるのは線形分離可能なクラスに対してのみである。<br>

オッズ比<br>
事象の起こりやすさを表している<br>
### $\frac{p}{(1-p)}$
ロジット関数<br>
オッズ比の対数<br>
### $logit(p) = log\frac{p}{(1-p)}$
ロジット関数は0 - 1の範囲の入力を受け取り、実数の全範囲に変換する。<br>
### $logit(p(y=1|\mathbf{x})) = w_{0}x_{0}+w_{1}x_{1}+…+w_{m}x_{m}=\Sigma_{i=0}^{m}w_{i}x_{i} = \mathbf{w^{T}x}$
ロジット関数の逆関数を、シグモイド関数という<br>
### $\phi(z)=\frac{1}{1+e^{-z}}$
zは総入力である。<br>
### $z = \mathbf{w^{T}x} = w_{0}x_{0}+w_{1}x_{1}+…+w_{m}x_{m}$

ロジスティック回帰モデルに対する直観を養うために、以下の図を参照にして説明をする。

```python
Image(filename='./images/03-00.png', width=500)
```

```python
Image(filename='./images/03-01.png', width=500)
```

特徴量$x$が重み$w$でパラメータ化されるとしたら、このシグモイド関数の出力は<br>
サンプルがクラス１に属している確率$\phi(z)=P(y=1|\mathbf{x};\mathbf{w})$であると解釈される<br>
あるサンプルに対して$\phi(z)=0.8$が算出されたら、このサンプルがクラス１に属する確率は80%である。<br>
同様に、２値分類と仮定すると、クラス2に分類される確率は$1-0.8=0.2$より、20%である。<br>
あとは、量子化器（単位ステップ関数）を使って予測された確率を２値の結果に変換する。<br>
### $\hat{y}= 1　\  \phi(z) > 0.5 \ $ or $ \ \hat{y} = 0 \  \phi(z) < 0.5$
シグモイド関数を参照すると以下と同じであることがわかる。<br>
### $\hat{y}= 1　\  z > 0.0 \ $ or $ \ \hat{y} = 0 \  z < 0.0$

### <font color='blue'>ロジスティック関数の重みの学習</font>

コスト関数<br>
### $J(\mathbf{w}) = \Sigma_{i}\frac{1}{2}(\phi(z^{(i)})-y^{(i)})^{2}$
を定義したとして、重み$\mathbf{w}$を学習するために、上記の関数を最小化した。<br>
ロジスティック回帰モデルの構築時に最大化したい尤度$L$を定義する。<br>
### $L(\mathbf{W})=P(y|\mathbf{x;w}) = \prod_{i=1}^nP(y^{(i)}| x^{(i)}; \mathbf{w}) = \prod_{i=1}^n(\phi(z^{(i)}))^{y(i)}(1-\phi(z^{(i)}))^{1-y^{(i)}}$
この式の自然対数をとると<br>
### $l(\mathbf{w}) = logL(\mathbf{w}) = \Sigma_{i=1}^{n}[y^{(i)}log(\phi(z^{(i)}))+(1-y^{(i)})log(1-\phi(z^{(i)}))]$
以上から、最適化アルゴリズムを用いて、対数尤度関数を最大化もしくは最小化できるようになった。<br>
ロジスティック関数のコスト関数を以下のように定義する。<br>
### $J(\mathbf{w}) = \Sigma_{i=1}^{n}[-y^{(i)}log(\phi(z^{(i)}))-(1-y^{(i)})log(1-\phi(z^{(i)}))]$
１つのサンプルの例から、ロジスティック関数のコスト関数についてみる。<br>
### $J(\phi(z), y;\mathbf{w}) = -ylog(\phi(z))-(1-y)log(1-\phi(z))$
上記から$y = 0$であれば、１つ目の項が0になり、$y = 1$であれば２つめの項が0になることがわかる。<br>

以下のグラフは、$\phi(z)$の様々な値に対する１つのサンプルの分類コストを表している。

```python
Image(filename='./images/03-02.png', width=500)
```

サンプルがクラス１に属していることを正しく理解した場合は、コストが０に近づき<br>
同様に、クラス０に属していることを正しく理解した場合は、コストが0に近づいてることがわかる。<br>
予測を間違えると、コストを引き上げることで、重みの学習をする。

## <font color='blue'>scikit-learnによるロジスティック回帰モデルのトレーニング</font>

```python
# ライブラリのロード
from sklearn.linear_model import LogisticRegression
```

```python
# ロジスティック回帰のインスタンスを生成
lr = LogisticRegression(C=1000.0, random_state=0)
```

```python
# トレーニングデータをモデルに適合
lr.fit(X_train_std, y_train)
```

```python
# 決定境界線をプロット
plot_decision_regions(X_combined_std, y_combined, classifier=lr, test_idx=range(105, 150))
plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.show()
```

## カーネルSVMを用いた非線形問題の求解

```python
# 乱数種を推定
np.random.seed(0)
```

```python
# 標準正規分布に従う乱数で200行2列の行列を生成
X_xor = np.random.randn(200, 2)
```

```python
# ２つの引数に対して排他的論理和を実行
y_xor = np.logical_xor(X_xor[:, 0] > 0, X_xor[:, 1] > 0)
```

```python
# 排他的論理和の値が真の時は1, 偽の時は-1を割り当てる
y_xor = np.where(y_xor, 1, -1)
```

```python
# ラベル１を青のxでプロット
plt.scatter(X_xor[y_xor==1, 0], X_xor[y_xor==1, 1], c='b', marker='x', label='1')
# ラベル-1を赤の四角でプロット
plt.scatter(X_xor[y_xor==-1, 0], X_xor[y_xor==-1, 1], c='r', marker='s', label='-1')
plt.xlim([-3, 3])
plt.ylim([-3, 3])
plt.legend(loc='best')
plt.show()
```

```python
# RBFカーネルによるSVMのインスタンスを生成
svm = SVC(kernel='rbf', random_state=0, gamma=0.10, C=10.0)
svm.fit(X_xor, y_xor)
plot_decision_regions(X_xor, y_xor, classifier=svm)
plt.legend(loc='upper left')
plt.show()
```

```python
# RBFカーネルによるSVMのインスタンスを生成（２つのパラメータを変更）
svm = SVC(kernel='rbf', random_state=0, gamma=0.2, C=1.0)
svm.fit(X_train_std, y_train)
plot_decision_regions(X_combined_std, y_combined, classifier=svm, test_idx=range(105, 150))
plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.show()
```

```python
# RBFカーネルによるSVMのインスタンスを生成（２つのパラメータを変更）
svm = SVC(kernel='rbf', random_state=0, gamma=100, C=1.0)
svm.fit(X_train_std, y_train)
plot_decision_regions(X_combined_std, y_combined, classifier=svm, test_idx=range(105, 150))
plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.show()
```
