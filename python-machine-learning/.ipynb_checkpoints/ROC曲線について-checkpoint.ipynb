ROC曲線を用いるのは、試験の点数から大学に合格(T)か不合格(F)の予測<br>
また、検査値から病気(T)か健康(F)かを判断したいときに用いる。<br>

```python
from IPython.display import Image
%matplotlib inline
Image(filename='./Images/ROC-0.png', width=100)
```

上記の表をみると、与えられた値のどこを切ってもTとFを完全に分離できないということがわかる<br>

ここで、11で切って10以上を陽性(positive)、11未満を陰性(negative)とした時<br>
全体の10個のTのうちの5個がpositiveに入るので、true positive(真陽性)の割合は0.5<br>
全体の5個のうちの1個がpositiveに入るので、false positive(偽陽性)の割合は0.2<br>

上記のように区切る値を色々と変化させて<br>
横軸にfalse positiveの割合、縦軸にtrue positiveの割合をとってプロットしたものがROC曲線である。<br>
