## <font color='blue'>PCAアルゴリズム（Principal Component Analysis）のアプローチについて</font>
1. d次元のデータセットを標準化<br>
2. 標準化したデータセットの共分散行列を作成<br>
3. 共分散行列を固有ベクトルと固有値に分解<br>
4. 最も大きいk個の固有値に対応するk個の固有ベクトルを選択。この場合のkは新しい特徴部分空間の次元数を表す<br>
5. 上位k個の固有ベクトルから射影行列Wを作成<br>
6. 射影行列Wを使ってd次元の入力データセットXを変換し、新しいk次元の特徴部分空間を取得<br>

## <font color='blue'>共分散行列の固有対を求める</font>

```python
import pandas as pd
import numpy as np

df_wine = pd.read_csv('https://archive.ics.uci.edu/ml/'
                      'machine-learning-databases/wine/wine.data',
                      header=None)

df_wine.columns = ['Class label', 'Alcohol', 'Malic acid', 'Ash',
                   'Alcalinity of ash', 'Magnesium', 'Total phenols',
                   'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins',
                   'Color intensity', 'Hue',
                   'OD280/OD315 of diluted wines', 'Proline']

df_wine.head()
```

```python
from sklearn.cross_validation import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# ２列目以降のデータをXに, 1列目のデータをyに格納
X, y = df_wine.iloc[:, 1:].values, df_wine.iloc[:, 0].values
# トレーニングデータとテストデータに分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# 平均と標準偏差を用いて標準化
sc = StandardScaler()
```

```python
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
import numpy as np
# 共分散行列を作成
cov_mat = np.cov(X_train_std.T)
cov_mat.shape
```

```python
# 固有値分解し, 固有値と固有ベクトルを作成
eigen_vals, eigen_vecs = np.linalg.eig(cov_mat)
```

```python
eigen_vals.shape
```

```python
eigen_vecs.shape
```

### <font color='blue'>固有値の分散説明率をプロット</font>

```python
# 固有値を合計
tot = sum(eigen_vals)
# 分散説明率を計算
var_exp = [(i / tot) for i in sorted(eigen_vals, reverse=True)]
# 分散説明率の累積和を取得
cum_var_exp = np.cumsum(var_exp)
import matplotlib.pyplot as plt
# 分散説明率の棒グラフを作成
plt.bar(range(1, 14), var_exp, alpha=0.5, align='center', label='individual explained variance')
# 分散説明率の累積和の階段グラフを作成
plt.step(range(1, 14), cum_var_exp, where='mid', label='comulative explained variance')
plt.ylabel('Explaned variance ratio')
plt.xlabel('Principal components')
plt.legend(loc='best')
plt.show()
```

#### <font color='red'><分散説明率のグラフに対する考察></font>
１つ目の主成分だけで分散の４０%近くを占めていることがわかる。<br>
また、最初の２つの主成分を合わせると、分散の６０％近くになっている<br>

### <font color='blue'>特徴変換</font> 

共分散行列を作成し, 固有ベクトルと固有値に分解できたところで, 最後の３つのステップに進み<br>
Wineデータセットを新しい主成分軸に変換する。<br>
<font color='red'>ここでは, 固有値の大きいものから順に固有対を並べ替え、選択された固有ベクトルから射影行列を生成する<br></font>

```python
# (固有値, 固有ベクトル)のタプルのリストを作成
eigen_pairs = [(np.abs(eigen_vals[i]), eigen_vecs[:, i]) for i in range(len(eigen_vals)) ]
```

```python
# (固有値, 固有ベクトル)のタプルを大きいものから順に並び替え
eigen_pairs.sort(key=lambda k: k[0], reverse=True)
```

次に、最も大きい２つの固有値に対応する２つの固有ベクトルを集める。<br>
それにより、このデータセットにおける分散の約６０％を捉えることができる<br>
実務では、計算効率と分類器の性能のバランスを見ながら、主成分の個数を決定する<br>

```python
len(eigen_pairs)
```

```python
eigen_pairs[0][0]
```

```python
len(eigen_pairs[0])
```

```python
len(eigen_pairs[0][1])
```

```python
# 上位２つの固有ベクトルから13×2次元の射影行列Wが作成される。
w = np.hstack((eigen_pairs[0][1][:, np.newaxis], eigen_pairs[1][1][:, np.newaxis]))
```

```python
print('Matrix W:\n', w)
```

上記の射影行列を用いてサンプルxをPCAの部分空間に変換し<br>
２つの新しい特徴量からなる２次元のサンプルベクトルを作成する

```python
X_train_std[0].dot(w)
```

同様に、行列の積を計算することで、124×13次元のトレーニングデータセット全体を２つの主成分に変換できる

```python
X_train_pca = X_train_std.dot(w)
```

```python
X_train_pca.shape
```

最後に、124×2次元の行列として格納された変換後のWineトレーニングデータセットを<br>
２次元の散布図としてプロットする。

```python
colors = ['r', 'b', 'g']
markers = ['s', 'x', 'o']
# 「クラスラベル」「点の色」「点の種類」の組み合わせからなるリストを生成してプロット
for l, c, m, in zip(np.unique(y_train), colors, markers):
    plt.scatter(X_train_pca[y_train==l, 0], X_train_pca[y_train==l, 1], c=c, label=l, marker=m)
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.legend(loc='lower left')
plt.show()
```

## <font color='blue'>scikit-learnの主成分分析</font>

```python
from matplotlib.colors import ListedColormap

def plot_decision_regions(X, y, classifier, resolution=0.02):

    # setup marker generator and color map
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    # plot class samples
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1],
                    alpha=0.8, c=cmap(idx),
                    edgecolor='black',
                    marker=markers[idx], 
                    label=cl)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA
```

```python
# 主成分数を指定して、PCAのインスタンスを作成
pca = PCA(n_components=2)
```

```python
# ロジスティック回帰のインスタンスを作成
lr = LogisticRegression()
```

```python
# トレーニングデータやテストデータをPCAに適合
X_train_pca = pca.fit_transform(X_train_std)
X_test_pca = pca.transform(X_test_std)
```

```python
# トレーニングデータをロジスティック回帰に適合
lr.fit(X_train_pca, y_train)
```

```python
plot_decision_regions(X_train_pca, y_train, classifier=lr)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend(loc='lower left')
plt.show()
```

```python
plot_decision_regions(X_test_pca, y_test, classifier=lr)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend(loc='lower left')
plt.show()
```

#### <font color='red'><考察></font>
ロジスティック回帰の性能が大幅に改善されていることがわかる。

## <font color='blue'>線形判別分析による教師ありデータ圧縮（LDA）</font>

LDAは特徴抽出手法の１つである。<br>
計算効率を高め、正則化されていないモデルで「次元の呪い」による過学習を抑制するために使用できる。<br>
PCAが分散の最も大きい直交成分を見つけるのに対し、LDAはクラスの分類を最適化する特徴部分空間を見つけ出す。<br>
PCAは教師なしで、LDAは教師ありである。<br>

以下の画像は２クラス問題に対するLDAの考え方である<br>

```python
from IPython.display import Image
%matplotlib inline

Image(filename='./images/05-01.png', width=400)
```

<font color='red'>画像に対する考察</font><br>
x軸(LD 1)に示されている線形判別により、正規分布に従う2つのクラスがうまく分割されているということがわかる。<br>
y軸(LD 2)に示されている典型的な線形判別は、データセットの分散の大半を捕捉するもの<br>
クラスの判別情報を全く捕捉しないため、よい線形判別とは言えない。<br>

LDAでは、データが正規分布に従っているということが前提となる。<br>
また、クラスの共分散行列が全く同じであることと、特徴量が統計的に見て互いに独立していることが前提である。<br>
<font color='red'>ただし、それらの前提を多少満たしていなくても、次元削減の手段としてのLDAはそれなりにうまくいく</font>

## <font color='blue'>LDA（Linear Discriminant Analysis）アルゴリズムのアプローチについて</font>
1. d次元データセットを標準化する。（dは特徴量の個数）
2. クラスごとにd次元の平均ベクトルを計算。
3.クラス間変動行列$S_{B}$と、クラス内変動行列$S_{W}$を生成。
4. 行列$S_{W}^{-1}$$S_{B}$の固有ベクトルと対応する固有値を計算する。
5. $d×k$次元の変換行列$W$を生成するために、最も大きいk個の固有値に対応する<br>
    k個の固有ベクトルを選択する。固有ベクトルは、この行列の列である。<br>
6. 変換行列${W}$を使ってサンプルを新しい特徴部分空間へ射影する。

### <font color='blue'>変動行列を計算する</font>

標準化はもうすんでいるので、平均ベクトルの計算に進む。<br>
そして、それらの平均ベクトルを使って、クラス内変動行列とクラス間変動行列を生成する。<br>
各平均ベクトル$m_{i}$には、クラス$i$のサンプルに関する平均特徴量の値$\mu_{i}$が格納される<br>

```python
np.set_printoptions(precision=4)

mean_vecs = []
for label in range(1, 4):
    mean_vecs.append(np.mean(X_train_std[y_train == label], axis=0))
    print('MV %s: %s\n' % (label, mean_vecs[label - 1]))
```

変動ベクトルを使ってクラス内変動行列$S_{w}$を計算する方法は以下のよう。<br>
$S_{w}=\sum_{i=1}^{c} S_{i}$ 

ここで$S_{i}$は以下のようになる。<br>
$S_{i} = \sum_{x \in D_{i}}(x - m_{i})(x - m_{i})$

### <font color='blue'>クラス内変動行列（共分散行列）の計算</font>

```python
d = 13
S_W = np.zeros((d, d))

# 最初にクラスを指定
for label,mv in zip(range(1,4), mean_vecs):
    
    class_scatter = np.zeros((d, d))
    # クラス内のベクトルを用いて、変動行列の合計を算出
    for row in X_train_std[y_train == label]:
        
        row, mv = row.reshape(d, 1), mv.reshape(d, 1)
        class_scatter += (row-mv).dot((row-mv).T)
    
    # クラス内変動行列の算出
    S_W += class_scatter
    
print('Within-class scatter matrix: %sx%s' % (S_W.shape[0], S_W.shape[1]))
```

<font color='red'>変動行列を計算するときには、トレーニングデータセットにおいてクラスラベルが一様に分布していることが前提。</font><br>
以下ではクラスラベルが一様に分布しているか確かめる。<br>

```python
print('Class label distribution: %s' 
      % np.bincount(y_train)[1:])
```

上記からクラスラベルが一様でないということがわかる。<br>
従って以下の式を用いて、スケーリングを行う必要がある<br>
### $\sum_{i} = \frac{1}{N_{i}}S_{w} = \frac{1}{N_{i}}\sum_{x \in D_{i}}(x - m_{i})(x - m_{i})^T$
<font color='red'>ここで、この計算が実は共分散行列の計算と同じであることがわかる。</font><br>
共分散行列は変動行列の正規化バージョンである。<br>
<font color='red'>従って以上よりクラス内変動行列（共分散行列）を以下に作成する。</font><br>

```python
d = 13 # 特徴量の個数
S_W = np.zeros((d, d))
for label, mv in zip(range(1, 4), mean_vecs):
    
    class_scatter = np.cov(X_train_std[y_train==label].T)
    S_W += class_scatter

print('クラス内変動行列: %sx%s' % (S_W.shape[0], S_W.shape[1]))
```

### <font color='blue'>クラス間変動行列$S_{B}$の作成</font>

$m$ : すべてのクラスのサンプルを対象として計算される全体平均<br>
$N_{i}$ : 各クラス内のサンプル数
### $S_{B} = \sum_{i}^{c}N_{i}(m_{i}-m)(m_{i}-m)^T$

```python
mean_overall = np.mean(X_train_std, axis=0) # すべてのクラスのサンプルを対象として計算される全体平均
d = 13 # 特徴量の個数
S_B = np.zeros((d, d))
for i, mean_vec in enumerate(mean_vecs):
    
    n = X_train[ y_train==i+1, : ].shape[0]
    mean_vec = mean_vec.reshape(d, 1)
    mean_overall = mean_overall.reshape(d, 1)
    S_B += n * (mean_vec - mean_overall).dot((mean_vec - mean_overall).T)
    
print('クラス間変動行列: %sx%s' % (S_B.shape[0], S_B.shape[1]))
```

### <font color='blue'>新しい特徴部分空間の線形判別を選択する</font>

LDAの残りのステップは、PCAのステップと似ている。<br>
ただし、共分散行列で固有値分解を実行するのでなく、行列$S_{w}^{-1}S_{B}$の一般化された固有値問題を解く

```python
# inv関数で逆行列, dot関数で行列積, eig関数で固有値を計算
eigen_vals, eigen_vecs = np.linalg.eig(np.linalg.inv(S_W).dot(S_B))
```

固有対を計算した後は、固有値を大きいものから降順で並び替えることができる。

```python
# Make a list of (eigenvalue, eigenvector) tuples
eigen_pairs = [(np.abs(eigen_vals[i]), eigen_vecs[:, i])
               for i in range(len(eigen_vals))]

# Sort the (eigenvalue, eigenvector) tuples from high to low
eigen_pairs = sorted(eigen_pairs, key=lambda k: k[0], reverse=True)

# Visually confirm that the list is correctly sorted by decreasing eigenvalues

print('Eigenvalues in decreasing order:\n')
for eigen_val in eigen_pairs:
    print(eigen_val[0])
```

線形判別（固有ベクトル）によってクラスの判別情報がどの程度捕捉されるかを測定<br>

```python
tot = sum(eigen_vals.real)
discr = [(i / tot) for i in sorted(eigen_vals.real, reverse=True)]
cum_discr = np.cumsum(discr)

plt.bar(range(1, 14), discr, alpha=0.5, align='center',
        label='individual "discriminability"')
plt.step(range(1, 14), cum_discr, where='mid',
         label='cumulative "discriminability"')
plt.ylabel('"discriminability" ratio')
plt.xlabel('Linear Discriminants')
plt.ylim([-0.1, 1.1])
plt.legend(loc='best')
plt.tight_layout()
# plt.savefig('./figures/lda1.png', dpi=300)
plt.show()
```

結果のグラフから分かるように、最初の２つの線形判別はWineトレーニングデータセット内の有益な情報をほぼ100%捕捉している。
<font color='red'>もっとも判別力のある２つの固有ベクトルを列方向に並べて、変換行列$W$を作成</font>

```python
w = np.hstack((eigen_pairs[0][1][:, np.newaxis].real,
              eigen_pairs[1][1][:, np.newaxis].real))
print('Matrix W:\n', w)
```

### <font color='blue'>新しい特徴空間にサンプルを射影する</font>

ここまで作成してきた変換行列$W$に行列をかけることにより、トレーニングデータセットを変換する。
### $X^{'} = XW$

```python
# 標準化したトレーニングデータに変換行列をかける
X_train_lda = X_train_std.dot(w)
```

```python
colors = ['r', 'b', 'g']
markers = ['s', 'x', 'o']

for l, c, m in zip(np.unique(y_train), colors, markers):
    plt.scatter(X_train_lda[y_train == l, 0] * (-1),
                X_train_lda[y_train == l, 1] * (-1),
                c=c, label=l, marker=m)

plt.xlabel('LD 1')
plt.ylabel('LD 2')
plt.legend(loc='lower right')
plt.tight_layout()
# plt.savefig('./figures/lda2.png', dpi=300)
plt.show()
```

<font color='red'><LDAによるクラス分けの考察></font>

<font color='red'>LDAによるクラス分けの結果の考察<br></font>
新しい特徴空間では、３つのクラスが完全に線形分離していることがわかる。

## <font color='blue'>カーネル主成分分析を使った非線形写像</font>

パーセプトロンを収束させるには、完全に線形分離可能なトレーニングデータが必要になる。<br>
<font color='red'>ここまで取り上げてきた他のアルゴリズムは、線形分離性が完全でないのはノイズのためであることを前提とする。<br></font>
だが、現実のアプリケーションではむしろ非線形問題に遭遇する方がはるかに多い。<br>
非線形問題に対処するためには、PCAやLDAといった次元削減方法は最良の選択肢ではないかもしれない<br>
そこでここでは、カーネルPCAを取り上げることにする。<br>

線形問題と非線形問題の違いを以下に示す。

## <font color='blue'>カーネル関数とカーネルトリック</font>

非線形問題を解くには、より<font color='red'>高い次元の新しい特徴空間へ射影</font>し、そこで線形分離可能な状態にする。<br>
### $\phi : {R}^{d} → R^{k} (k >> d) $

2次元空間を３次元空間に写像する例<br>
### $x = [x_{1}, x_{2}]^T \xrightarrow{\phi} z = [x_{1}^2, \sqrt{2x_{1}x_{2}},  x_{2}^2]^T$

上記の高次元空間に対して標準のPCAを適用し、より低次元の空間へ再び射影し、サンプルを線形分類器で分離できるようにする。<br>
<font color='red'>だがこの方法には、計算コストが非常に高いという問題点が存在している。</font><br>
そこでカーネルトリックの出番となる。<br>
カーネルトリックにより高次元空間に写像することなく、元の特徴空間において２つの高次元の特徴ベクトルの類似度を計算可能。<br>

ここで、「標準」のPCAアルゴリズムについてもう１度見てみる。
２つの特徴量$j$と$k$の間の共分散を以下のように計算。
### $\sigma_{jk} = \frac{1}{n} \Sigma_{i=1}^{n}(x_{j}^{(i)}-\mu_{j})(x_{k}^{(i)}-\mu_{k})$
平均値は$0$が中心なので
### $\mu_{j} = \frac{1}{n} \Sigma_{i} x_{j}^{(i)}=0$
よって
### $\sigma_{jk} = \frac{1}{n} \Sigma_{i=1}^{n}x_{j}^{i}x_{k}^{i}$
したがって一般化した共分散行列は以下のよう
### $\Sigma =\frac{1}{n}\Sigma_{i=1}^{n}\mathbf{x}^{(i)} (\mathbf{x}^{(i)})^{T}$
$\phi$を用いて非線形の特徴量の組み合わせに置き換えると
### $\Sigma = \frac{1}{n}\Sigma_{i=1}^{n}\phi(\mathbf{x}^{(i)}) \phi(x^{(i)})^{T}  　- ①$
この共分散から固有ベクトル（主成分）を取り出すには、以下の方程式を解く
### $\Sigma\mathbf{v} = \lambda \mathbf{v}  -②$
<font>①を②の左辺に代入して</font>
###  $\frac{1}{n}\Sigma_{i=1}^{n}\phi(\mathbf{x}^{(i)})\phi(\mathbf{x}^{(i)})^{T}\mathbf{v} = \lambda \mathbf{v}$
右辺を$\mathbf{v}について解くと$
### $\mathbf{v}=\frac{1}{n\lambda}\sum_{i=1}^n\phi(\mathbf{x}^{(i)})\phi(\mathbf{x}^{(i)})^T\mathbf{v}=\frac{1}{n}\sum_{i=1}^{n}\mathbf{a}^{(i)}\phi(\mathbf{x}^{(i)})$
$\lambda, \mathbf{v}$ : 固有値と固有ベクトル, $\mathbf{a}$ : カーネル行列$\mathbf{K}$の固有ベクトルを抽出<br>
カーネル行列を導出する方法は以下のよう
### $\sum = \frac{1}{n}\sum_{i=1}^{n}\phi(\mathbf{x}^{(i)})\phi(\mathbf{x}^{(i)})^T = \frac{1}{n}\phi(\mathbf{X})^T\phi(\mathbf{X})  -③$
これにより、固有ベクトル方程式は<font color='red'>式②より</font>以下のように記述できる
### $\mathbf{v}=\frac{1}{n}\sum_{i=1}^{n}\mathbf{a}^{(i)}\phi(\mathbf{x}^{(i)})=\lambda\phi(\mathbf{X})^T\mathbf{a}  -④$
ここで$\sum\mathbf{v}=\lambda\mathbf{v}$であるため、<font color='red'>式③と式④から</font>以下のように記述できる
### $\sum\mathbf{v} = \frac{1}{n}\phi(\mathbf{x})^T\phi(\mathbf{x})\mathbf{v}=\frac{1}{n}\phi(\mathbf{x})^T\phi(\mathbf{x})\lambda\phi(\mathbf{x})^T\mathbf{a}  -⑤$
ここで、$\sum\mathbf{v}=\lambda\mathbf{v}$の右辺に$\mathbf{v}=\lambda\phi(\mathbf{x})^T\mathbf{a}$を代入すると
### $\lambda\mathbf{v}=\lambda^2\phi(\mathbf{x})^T\mathbf{a}  -⑥$
$⑤ = ⑥$より
### $\frac{1}{n}\phi(\mathbf{x})^T\phi(\mathbf{x})\lambda\phi(\mathbf{x})^T\mathbf{a} = \lambda^2\phi(\mathbf{x})^T\mathbf{a}$ 
$\lambda$で割って
### $\frac{1}{n}\phi(\mathbf{X})^T\phi(\mathbf{X})\phi(\mathbf{X})^T\mathbf{a} = \lambda\phi(\mathbf{X})^T\mathbf{a}  -⑦$
③の両辺に左から$\phi(\mathbf{x})$を掛けると、以下のようになる
### $\frac{1}{n}\phi(\mathbf{X})\phi(\mathbf{X}^T)\phi(\mathbf{X})\phi(\mathbf{X}^T)\mathbf{a}=\lambda\phi(\mathbf{X})\phi(\mathbf{X})^T\mathbf{a}$
### $⇒\frac{1}{n}\phi(\mathbf{X})\phi(\mathbf{X})^T\mathbf{a}=\lambda\mathbf{a}  -⑧$
### $⇒\frac{1}{n}\mathbf{Ka}=\lambda\mathbf{a}$
ここで、$\mathbf{K}$は類似度（カーネル）行列である。<br>
式⑧の右辺の$\phi(\mathbf{X})\phi(\mathbf{X})^T$を以下のように置換できる
### $\mathbf{K} = \phi(\mathbf{X})\phi(\mathbf{X})^T$
上記からカーネルトリックを使用することで、<font color='red'>サンプル$\mathbf{x}$の関数$\phi$どうしの内積の計算をカーネル関数$K$により回避</font><br>
それにより、固有ベクトルを明示的に計算する必要がなくなる。
### $k(\mathbf{x}^{(i)}, \mathbf{x}^{(j)}) = \phi(\mathbf{x}^{(i)})^T\phi(\mathbf{x}^{(j)})$
<font color='red'>標準のPCAのアプローチとは異なり、変換行列は生成しない。</font><br>
<font color='red'>基本的に、カーネル関数については、２つのベクトル間の内積、つまり類似度の目安を計算する関数</font><br>

## <font color='blue'>よく使用されるカーネルは以下のよう</font>

・多項式カーネル。$\theta$はしきい値、$P$はユーザーが指定しなければならない指数<br>
### $k(\mathbf{x}^{(i)}, \mathbf{x}^{(j)}) = (\mathbf{x}^{(i)T}\mathbf{x^{(j)}}+\theta)$
・双曲線正接（S字）カーネル
### $k(\mathbf{x}^{(i)}, \mathbf{x}^{(j)}) = \tanh(\eta\mathbf{x}^{(i)T}\mathbf{x}^{(j)}+\theta)$
・動径基底関数（RBF）またはガウスカーネル。
### $k(\mathbf{x^{(i)}, \mathbf{x^{(j)}}})=exp(-\gamma||\mathbf{x}^{(i)}-\mathbf{x}^{(j)}||^2)$

## <font color='blue'>以上の手順のまとめ</font>

1. カーネル（類似度）行列$K$を計算し、以下の計算を行う<br>
### $k(\mathbf{x}^{(i)}, \mathbf{x}^{(j)}) = exp(-\gamma||\mathbf{x}^{(i)}-\mathbf{x}^{(j)}||^2)$
この計算をサンプルごとに行う。<br>
### $A = \left(\begin{array}{ccc}k(
\mathbf{x}^{(1)}, \mathbf{x}^{(1)}) & k(\mathbf{x}^{(1)}, \mathbf(x)^{(2)}) & ... & k(\mathbf{x}^{(1)}, \mathbf{x}^{(n)}) 
\\ k(\mathbf{x}^{(2)}, \mathbf{x}^{(1)}) & k(\mathbf{x}^{(2)}, \mathbf{x}^{(2)}) & ... & k(\mathbf{x}^{(2)}, \mathbf{x}^{(n)})
\\ ... & ... & ... & ...
\\ k(\mathbf{x}^{(n)}, \mathbf{x}^{(1)}) & k(\mathbf{x}^{(n)}, \mathbf{x}^{(2)}) & ... & k(\mathbf{x}^{(n)}, \mathbf{x}^{(n)})
\end{array}
\right)$
トレーニングサンプルが１００個含まれている場合、ペアごとの類似度からなる対称カーネル行列は$100×100$次元になる。<br>
2. 以下の式を用いてカーネル行列$K$の中心化を行う。
### $K^{'}=K-1_{n}K-K1_{n}+1_{n}K1_{n}$
$1_{n}$はすべての値が$\frac{1}{n}$に等しい$n×n$次元の行列である。<br>
3. 対応する固有値に基づき、中心化されたカーネル行列の$k$個の固有ベクトルを収集する。<br>
この場合、固有値は大きい順にランク付けされている。<br>
標準のPCAとは対称的に、固有ベクトルは主成分でなく、それらの軸に射影されるサンプルである。<br>

### <font color='blue'>PCA、LDA、カーネルトリックについて</font>

### <font color='blue'>PCA、LDAの違いについて</font>
どちらも最初にデータセットを標準化している。<br>
<font color='red'>PCAでは、共分散行列の固有値、固有ベクトルを計算。</font><br>
<font color='red'>LDAではクラス間変動行列$S_{B}$とクラス内変動行列$S_{w}$の行列$S_{w}^{-1}S_{B}$の固有ベクトルと対応している固有値を計算。</font><br>

### <font color='blue'>カーネル関数とカーネルトリックについて</font>
最初に共分散行列から固有値、固有ベクトルを取り出すようにしているが<br>
<font color='red'>固有値ベクトルを取り出す過程の式を変換し、カーネル（類似度）行列$K$の固有ベクトルを抽出。</font><br>

## <font color='blue'>Pythonでカーネル主成分分析を実装する</font>

```python
from scipy.spatial.distance import pdist, squareform
from scipy import exp
from scipy.linalg import eigh
import numpy as np

def rbf_kernel_pca(X, gamma, n_components):
    """
    RBF kernel PCA implementation.

    Parameters
    ------------
    X: {NumPy ndarray}, shape = [n_samples, n_features]
        
    gamma: float
      Tuning parameter of the RBF kernel
        
    n_components: int
      Number of principal components to return

    Returns
    ------------
     X_pc: {NumPy ndarray}, shape = [n_samples, k_features]
       Projected dataset   

    """
    # M×N次元のデータセットでペアごとの平方ユークリッド距離を計算
    sq_dists = pdist(X, 'sqeuclidean')

    # ペアごとの距離を正方行列に変換
    mat_sq_dists = squareform(sq_dists)

    # 対称カーネル行列を計算
    K = exp(-gamma * mat_sq_dists)

    # カーネル行列を中心化
    N = K.shape[0]
    one_n = np.ones((N, N)) / N
    K = K - one_n.dot(K) - K.dot(one_n) + one_n.dot(K).dot(one_n)

    # 中心化されたカーネル行列から固有対称を取得
    eigvals, eigvecs = eigh(K)

    # 上位k個の固有ベクトル（射影されたサンプル）を収集
    X_pc = np.column_stack((eigvecs[:, -i]
                            for i in range(1, n_components + 1)))

    return X_pc
```

## <font color='blue'>新しいデータ点を射影する</font>

実際の業務では、モデルを構築した後に、新しいサンプルも収集する。<br>
ここでは、トレーニングデータセットに含まれていなかったデータ点を射影する方法について説明する。<br>

主成分軸に新しいサンプル$\mathbf{x^{'}}$を射影したい場合は、以下の式を計算する必要性がある。<br>
### $\phi(\mathbf{x^{'}})^{T}\mathbf{v}$
ここで、$\mathbf{v}$は主成分軸である。<br>
だが、カーネルトリックを利用すると、射影$\phi(\mathbf{x^{'}})^{T}\mathbf{v}$を明示的に計算しなくとも大丈夫。<br>
ただし、カーネルPCAはメモリベースの手法であり、新しいサンプルを射影するたびに、元のトレーニングデータセットを再利用する必要がある。<br>
トレーニングデータセットの$i$番目のサンプルと、新しいサンプル$\mathbf{x^{'}}$の間でペアごとのRBFカーネルを計算する。<br>
### $\phi(\mathbf{x^{'}})^{T}\mathbf{v} = \Sigma_{i}\mathbf{a^{(i)}}\phi{(x^{'})^{T}}\phi{(x^{(i)})}$
### $=\Sigma_{i}\mathbf{a}^{(i)}k(\mathbf{x}^{'}, \mathbf{x}^{(i)})^{T}$
ここで、カーネル行列$K$の固有ベクトル$a$と固有値$\lambda$は、以下の方程式を満たしている。
### $\mathbf{K}\mathbf{a} = \lambda\mathbf{a}$

```python
from scipy.spatial.distance import pdist, squareform
from scipy import exp
from scipy.linalg import eigh
import numpy as np

def rbf_kernel_pca(X, gamma, n_components):
    """
    RBF kernel PCA implementation.

    Parameters
    ------------
    X: {NumPy ndarray}, shape = [n_samples, n_features]
        
    gamma: float
      Tuning parameter of the RBF kernel
        
    n_components: int
      Number of principal components to return

    Returns
    ------------
     X_pc: {NumPy ndarray}, shape = [n_samples, k_features]
       Projected dataset   
     
     lambdas: list
       Eigenvalues

    """
    # Calculate pairwise squared Euclidean distances
    # in the MxN dimensional dataset.
    sq_dists = pdist(X, 'sqeuclidean')

    # Convert pairwise distances into a square matrix.
    mat_sq_dists = squareform(sq_dists)

    # Compute the symmetric kernel matrix.
    K = exp(-gamma * mat_sq_dists)

    # Center the kernel matrix.
    N = K.shape[0]
    one_n = np.ones((N, N)) / N
    K = K - one_n.dot(K) - K.dot(one_n) + one_n.dot(K).dot(one_n)

    # Obtaining eigenpairs from the centered kernel matrix
    # numpy.eigh returns them in sorted order
    eigvals, eigvecs = eigh(K)

    # Collect the top k eigenvectors (projected samples)
    alphas = np.column_stack((eigvecs[:, -i]
                              for i in range(1, n_components + 1)))

    # Collect the corresponding eigenvalues
    lambdas = [eigvals[-i] for i in range(1, n_components + 1)]

    # 上位k個の固有ベクトル, 対応する固有値
    return alphas, lambdas
```

### <font color='blue'>半月形データセットにRBFカーネルPCAを当てはめる</font>

```python
# データセットの作成
X, y = make_moons(n_samples=100, random_state=123)
alphas, lambdas = rbf_kernel_pca(X, gamma=15, n_components=1)
```

半月型データセットから26番目の点は新しいデータ点$\mathbf{x}^{'}$であり、このデータ点を新しい部分空間へ射影すると仮定する

```python
x_new = X[25]
x_new.shape, x_new
```

```python
x_proj = alphas[25] # 元の射影されたベクトル
x_proj.shape, x_proj
```

```python
def project_x(x_new, X, gamma, alphas, lambdas):
    pair_dist = np.array([np.sum((x_new - row)**2) for row in X])
    k = np.exp(-gamma * pair_dist)
    return k.dot(alphas / lambdas)
```

```python
x_reproj = project_x(x_new, X, gamma=15, alphas=alphas, lambdas=lambdas)
```

```python
x_reproj.shape, x_reproj
```

上記から新しい点と仮定された点は、元の点と同じところに射影された

## <font color='blue'>PCA、LDA、カーネルトリックについて</font>
### <font color='blue'>PCA、LDAの違いについて</font>
どちらも最初にデータセットを標準化している。<br>
<font color='red'>PCAでは、共分散行列の固有値、固有ベクトルを計算。</font><br>
<font color='red'>LDAではクラス間変動行列$S_{B}$とクラス内変動行列$S_{w}$の行列$S_{w}^{-1}S_{B}$の固有ベクトルと対応している固有値を計算。</font><br>
### <font color='blue'>カーネル関数とカーネルトリックについて</font>
最初に共分散行列から固有値、固有ベクトルを取り出すようにしているが<br>
<font color='red'>固有値ベクトルを取り出す過程の式を変換し、カーネル（類似度）行列$K$の固有ベクトルを抽出。</font><br>

## <font color='blue'>さまざまな前処理の手法を効率よく組み合わせ</font><br>
## <font color='blue'>さまざまなモデルのベストプラクティスを取得できる。</font><br>
